```python
# Importamos todas las librerías necesarias para nuestro programa

import tweepy
import numpy as np
import pandas as pd
import sys
import jsonpickle
import os

# Comprobaciones de estilo PEP-8 con: %%pycodestyle
# %load_ext pycodestyle_magic
```

```python
# Nos autenticamos en la aplicación con las claves proporcionadas por Twitter

consumer_key =  # Introducir consumer key
consumer_secret =  # Introducir consumer secret

token =  # Introducir token
secret =  # Introducir secret

# Utilizamos esta autenticación para conseguir mayor rapidez de descarga
auth = tweepy.AppAuthHandler(consumer_key, consumer_secret)
# auth.set_access_token(token, secret)

# Indicamos a la aplicación que se pause cuando se alcanze el rate limit
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# Definimos todas las variables que influyen en la recogida de datos

searchQuery = 'Andalucia'  # Término a buscar
maxTweets = 150000  # Cantidad máxima de tweets a descargar
tweetsPerQry = 100  # Cantidad de tweets por consulta (Máx = 100)
fName = 'tweets.txt'  # Nombre del archivo donde guardar tweets (.txt)
sinceId = None  # Indicamos una ID de tweet desde la cual buscar
max_id = 0  # Guarda el ID del último tweet descargado
tweetCount = 0  # Número de tweets descargados hasta el momento
desde = "2018-12-01"  # Fecha desde la que buscar tweets (Máx. 1 semana)
idioma = "es"  # Idioma de los tweets en la búsqueda
# Añadimos la siguiente línea para que en la búsqueda no aparezcan RT
searchQuery = searchQuery + ' -filter:retweets'
```

```python
# Creamos el data frame de salida mediante pandas con los campos requeridos
output = pd.DataFrame({'usuario': [],
                       'texto': [],
                       'fecha': []})
```

```python
# Fragmento de código extraído de la implementación de la siguiente URL:
# https://bhaskarvk.github.io/2015/01/how-to-use-twitters-search-rest-api-most-effectively./
# Modificado para adaptarlo a nuestro caso de uso de generación del data frame

print("Descargando max {0} tweets".format(maxTweets))
with open(fName, 'w') as f:
    # Bucle para descargar el número de tweets indicados.
    while tweetCount < maxTweets:
        try:
            if(max_id <= 0):
                # Realizamos la búsqueda inicial (máx_id == 0)
                if(not sinceId):
                    new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                            lang=idioma, since=desde,
                                            tweet_mode='extended')
                else:
                    new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                            lang=idioma, since=desde,
                                            tweet_mode='extended')
            else:
                # Búsquedas sucesivas cuando (máx_id != 0)
                if (not sinceId):
                    new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                            lang=idioma, since=desde,
                                            max_id=str(max_id - 1),
                                            tweet_mode='extended')
                else:
                    new_tweets = api.search(q=searchQuery, count=tweetsPerQry,
                                            lang=idioma, since=desde,
                                            max_id=str(max_id - 1),
                                            since_id=sinceId,
                                            tweet_mode='extended')
            if not new_tweets:
                # Detenemos la ejecución cuando no hay más tweets
                print("No se han encontrado más tweets")
                break
            for tweet in new_tweets:
                # Agregamos los campos necesarios al data frame para cada tweet
                output = output.append(pd.Series({'usuario': tweet.user.name,
                                                  'texto': tweet.full_text,
                                                  'fecha': tweet.created_at},
                                       name=tweet.id))
                # Escribimos el JSON del tweet en una nueva línea del .txt
                f.write(jsonpickle.encode(tweet._json, unpicklable=False)
                        + '\n')
            # Incrementamos el número de tweets en la cantidad de tweets
            # descargados por new_tweets
            tweetCount += len(new_tweets)
            print("Descargados {0} tweets".format(tweetCount))
            # Establecemos el nuevo max_id = id del último tweet descargado
            max_id = new_tweets[-1].id
        except tweepy.TweepError as e:
            # Si se produce algún error, detenemos la ejecución e
            # imprimimos el error
            print("Se ha producido algún error : " + str(e))
            break

print("Descargados {0} tweets, Guardados en {1}".format(tweetCount, fName))
```

```python
def cuenta_palabras(frase):
    """Determina el número de palabras de una frase"""
    # Construimos un vector de palabras separando el texto por espacios.
    auxiliar = frase.split(" ")
    # Devolvemos la longitud (número de palabras) del vector construido
    return(len(auxiliar))
```

```python
# Creamos las columnas adicionales en el data frame
output['palabras'] = output.texto.apply(cuenta_palabras)
# Contamos la longitud del texto del tweet mediante len
# Aunque los tweets, como máximo, cuentan con 280 caracteres,
# aparecen tweets de mayor tamaño, ya que lo que sigue a una @
# no cuenta para el límite de caracteres del tweet.
output['caracteres'] = output.texto.apply(len)
```

```python
def get_hour(h):
    """Retorna la hora de la fecha del tweet"""
    return(h.hour)
```

```python
# Creamos la columna 'hora' del data frame aplicando la función get_hour
output['hora'] = output.fecha.apply(get_hour)
output
```

```python
# Agrupamos el dataframe por la columna hora
output_agrupado = output.groupby(['hora'])
# Mediante la siguiente línea, contamos el número de tweets por hora
output_agrupado.count()['texto']
```

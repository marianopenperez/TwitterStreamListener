```python
# Importamos todas las librer√≠as necesarias para nuestro programa

import tweepy
import pandas as pd
import sys
import jsonpickle
import os
import time

# Comprobaciones de estilo PEP-8 con: %%pycodestyle
# %load_ext pycodestyle_magic
```

```python
# Nos autenticamos en la aplicaci√≥n con las claves proporcionadas por Twitter
consumer_key =  # Introducir consumer_key
consumer_secret =  # Introducir consumer_secret

token =  # Introducir token
secret =  # Introducir secret

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(token, secret)

# Indicamos a la aplicaci√≥n que se pause cuando se alcanze el rate limit
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
# Nombre del fichero donde almacenaremos los tweets en formato JSON
fName = 'streaming.txt'
# Tema o temas sobre los que haremos streaming
track_topic = 'Estrasburgo'
tweet_limit = 150000  # L√≠mite de tweets que queremos escuchar
time_limit = 10800  # Tiempo m√°ximo, en segundos, que queremos escuchar
```

```python
f = open(fName, 'w')  # Apertura del fichero en modo escritura
```

```python
# Creamos la clase MyStreamListener con la que escucharemos los tweets


class MyStreamListener(tweepy.StreamListener):

    # Definimos los atributos de la clase
    def __init__(self, api=None):
        self.api = api or API()
        self.counter = 0  # Contador de tweets le√≠dos
        self.time = time.time()  # Guarda el momento de inicio del streaming

    # Definimos el proceso de lectura de tweets
    def on_status(self, status):
        # Eliminamos retweets del streaming mediante las siguientes cl√°usulas
        # status.rewtweet_count == 0 (El tweet no tiene retweets)
        # not status.text.startswith('RT') (Elimina retweets manuales)
        if (status.retweet_count == 0) and (not status.text.startswith('RT')):
            # Intentamos imprimir el texto del tweet en modo extendido por si
            # encontramos tweets de m√°s de 140 caracteres.
            try:
                # Imprimimos por pantalla el n√∫mero de tweets le√≠dos
                # y el contenido del √∫ltimo tweet
                print("{0} ".format(self.counter+1) +
                      status.extended_tweet["full_text"])
                # Impresi√≥n del tiempo de ejecuci√≥n del streaming
                print("Streaming abierto desde hace " +
                      str(time.time()-self.time) + " segundos")
            # Si el tweet tiene menos de 140 caracteres,
            # no hay atributo extended_tweet y provoca el error AttributeError
            # en cuyo caso, manejamos el tweet con la siguiente cla√∫sula
            except AttributeError:
                # Imprimimos por pantalla el n√∫mero de tweets le√≠dos
                # y el contenido del √∫ltimo tweet
                print("{0} ".format(self.counter+1) + status.text)
                # Impresi√≥n del tiempo de ejecuci√≥n del streaming
                print("Streaming abierto desde hace " +
                      str(time.time()-self.time) + " segundos")

            # Escribimos el JSON del tweet en una nueva l√≠nea del .txt
            f.write(jsonpickle.encode(status._json, unpicklable=False) + '\n')
            self.counter += 1  # Incrementamos en uno los tweets le√≠dos
            # Definimos las condiciones de fin de streaming
            # Hemos le√≠do el n√∫mero de tweets indicado en tweet_limit
            # Hemos superado el tiempo de streaming indicado en time_limit
            if((self.counter >= tweet_limit)
               or ((time.time() - self.time) >= time_limit)):
                f.close()  # Cerramos el fichero de tweets
                if(f.closed is True):  # Comprobaci√≥n de cierre de fichero
                    print("Fichero cerrado correctamente.")
                return False  # Indica fin de streaming

    def on_error(self, status_code):
        if status_code == 420:
            return False  # Indica fin de streaming
```

```python
# Instanciamos un objeto de tipo MyStreamListener con nuestra API
myStreamListener = MyStreamListener(api)
# Instanciamos un objeto de tipo Stream con nuestras credenciales
# y el objeto de tipo MyStreamListener
myStream = tweepy.Stream(auth=api.auth, listener=myStreamListener)
```

```python
myStream.filter(track=[track_topic])  # Iniciamos el streaming
```

```python
myStream.disconnect()  # Desconectamos el streaming
print("Streaming desconectado")
```
